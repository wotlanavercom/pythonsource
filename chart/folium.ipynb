```python
# 마커 클러스터
import numpy as np

# uniform(low=35.5,high=37.5,size=100) : 최소값 35.5 ~ 최대 37.5 사이의 균등하게 100개 생성

data = np.array(
    [
        np.random.uniform(low=35.5,high=37.5,size=100),
        np.random.uniform(low=127,high=129,size=100)
    ]
).T
popups = [str(i) for i in range(100)]

m = folium.Map([36.5,128], zoom_start=8)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m
```

# pip install vincent
- Vega() 차트 표시할 때 차트 작성해주는 라이브러리

```python
import vincent
import json

scatter_points = {
    "x" : np.random.randn(50).cumsum(),
    "y" : np.random.randn(50).cumsum(),
}

scatter_chart = vincent.Scatter(scatter_points,iter_idx="x", width=400, height=200)
scatter_json = scatter_chart.to_json()
scatter_dict = json.loads(scatter_json)

m = folium.Map([36.5,128],zoom_start=7)

# Vega() : 지도 팝업 안에 차트 표시
popup = folium.Popup()
folium.Vega(scatter_json,height=250, width=450).add_to(popup)
folium.Marker([37,128],popup=popup).add_to(m)

popup = folium.Popup()
folium.Vega(scatter_dict,height=250, width=450).add_to(popup)
folium.Marker([36.5,127],popup=popup).add_to(m)

m
```

```python
url = "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"

# 미국 위,경도 => 다운로드함
# state_geo = requests.get(f"{url}/us-states.json")

# 실업률 정보
state_unemployment = f"{url}/US_Unemployment_Oct2012.csv"

statr_data = pd.read_csv(state_unemployment)
statr_data.head()
```

```python
from branca.colormap import linear

colormap = linear.YlGnBu_09.scale(
    statr_data.Unemployment.min(),
    statr_data.Unemployment.max(),
)

colormap
```

```python
# 지도 색상별로 실업률 표시
# folium.GeoJson + branca

statr_data_dict = statr_data.set_index("State")["Unemployment"]

m = folium.Map([43,-100],zoom_start=4)

folium.GeoJson(
    "./us-states.json",
    name="unemployment",
    style_function=lambda feature:{
        "fillColor":colormap(statr_data_dict[feature["id"]]),
        "color":"black",
        "weight":1,
        "dashArray":"5.5",
        "fillOpacity":0.6
    }
).add_to(m)

folium.LayerControl().add_to(m)
m
```

```python
# 지도 색상별로 실업률 표시
# folium.Choropleth

m = folium.Map([43,-100],zoom_start=4)

folium.Choropleth(
    geo_data= "./us-states.json",
    data=statr_data,
    columns=["State","Unemployment"],
    key_on="feature.id",
    fill_color="YlGnBu",
    fill_opacity=0.6,
    line_opacity=0.2,
    legend_name="Unemploument Rate(%)"
).add_to(m)

folium.LayerControl().add_to(m)
m
```

```python
url = "https://raw.githubusercontent.com/suanlab/dataset/master"

seoul_geo_url = f"{url}/seoul_municipalities_geo_simple.json"
seoul_geo = requests.get(seoul_geo_url).json()
seoul_geo
```

```python
seoul_population = f"{url}/seoul_population.csv"
seoul_data = pd.read_csv(seoul_population,encoding="utf-8")
seoul_data.head()
```

```python
m = folium.Map([37.52843,126.980238],zoom_start=10)

folium.GeoJson(
    seoul_geo,
    name="seoul_municipalities"
).add_to(m)

m
```

```python
# 인구수에 맞는 컬러바 생성

colormap = linear.Blues_09.scale(
    seoul_data.population.min(),
    seoul_data.population.max(),
)

colormap
```

```python
# 구명 인덱스 설정
population_ditc = seoul_data.set_index("name")["population"]
population_ditc.head()
```

```python
# 구 별 컬러 지정

color_ditc = { str(key):colormap(population_ditc[key])for key in population_ditc.keys()}
color_ditc
```

```python
# 구 별 인구수 색상별로 표현

m = folium.Map([37.52843,126.980238],zoom_start=10)

folium.GeoJson(
    seoul_geo,
    name="population",
    style_function=lambda feature:{
        "fillColor":color_ditc[feature["properties"]["name"]],
        "color":"black",
        "weight":1,
        "dashArray":"5.5",
        "fillOpacity":0.6
    }
).add_to(m)

colormap.add_to(m)

folium.LayerControl().add_to(m)
m
```
