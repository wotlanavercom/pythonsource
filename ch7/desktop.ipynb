```python
# 전체 스크린샷
p.screenshot("screeen.png")
```

```python
# 특정위치의 스크린 샷
# region(x축상단,y축상단, width, height)
p.screenshot("region.png",region=(30,30,250,900))
```

```python
# locateOnScreen : 이미지의 화면상 좌표 구하기

screen_locate = p.locateOnScreen("screeen.png",confidence=0.9)
print(screen_locate)
p.click(screen_locate)
```

```python
p.screenshot("file.png",region=(38,6,30,20))
```

```python
file_locate = p.locateOnScreen("file.png",confidence=0.9)
print(file_locate)
p.click(file_locate)
```

```python
# confidence 신뢰도 0.8 ~ 0.9
checkbox_locate = p.locateOnScreen("file.png",confidence=0.9)
print(checkbox_locate)
p.click(checkbox_locate)
```

```python
for i in p.locateAllOnScreen("checkbox.png"):
    print(i)
    p.click(i)
```

```python
p.getWindowsWithTitle("제목 없음")[0]

print(w)

if not w.isActive:
    w.activate()
```

```python
# pip install pyperclip

import pyperclip

# 메모장 가져오기
w = p.getWindowsWithTitle("제목 없음")[0]
# 메모장 활성화
pyperclip.copy("파이썬")

# 클립보드 내용 붙여넣기
p.hotkey("ctrl","v")
```

```python
def word_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
    
```

```python
word_write('abcdefgfjf')
word_write("가나다라마바사")
```
