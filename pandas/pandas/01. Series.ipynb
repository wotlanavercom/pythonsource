### 생성

#### 1) 리스트

```python
list1 = [70,80,90]
```

```python
import pandas as pd

ser = pd.Series(list1)
ser
```

```python
ser1 = pd.Series(list1,index=["국어","영어","수학"])
ser1
```

#### 2) 딕셔너리

```python
ser2 = pd.Series({"a":1,"b":2,"c":3})
ser2
```

#### Read

```python
# index 조회

ser1.index
```

```python
# values 조회
ser1.values
```

```python
# values 에서 상위 5개만 조회 : head(), 개수 지정 가능 head(3)
ser1.head(2)
```

```python
# values 에서 끝 5개만 조회 : tail(), 개수 지정 가능 tail(3)
ser1.tail(2)
```

```python
# shape : 데이터 구조 (행,열)
ser1.shape
```

#### Update


```python
# index 변경

ser1.index = ["미술","음악","체육"]
ser1
```

```python
# 데이터 타입 변경 : astype("변경할 타입") => 원본 변경 안됨

ser1 = ser1.astype("float64")
```

```python
ser1
```

```python
ser3 = pd.Series(["1.1","2.2","3.3"])
ser3
```

```python
ser3 = ser3.astype("float64")
ser3
```

```python
ser4 = pd.Series(["test","1.1","2.2","3.3"])
ser4
```

```python
ser4.astype("float64")
```

```python
# 특정 값 변경
ser1["음악"] = 85
ser1
```

```python
ser1 = ser1.rename("점수")
ser1
```

#### 삭제

```python
del ser1["체육"]
```

```python
ser1
```

#### 3) ndarray

```python
# numpy 데이터 타입 : ndarray

import numpy as np

ser5 = pd.Series(np.random.randn(4), index = list("abcd"))
ser5
```

#### 4) DataFrame 이용

```python
df = pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
df
```

```python
ser5 = pd.Series(df['a'])
ser5
```
