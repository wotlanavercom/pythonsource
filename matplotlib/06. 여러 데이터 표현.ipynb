## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
x = [1,2,3]
y = [2,4,8]     # 만명
data1 = [5,1,3]
data2 = [1,2,5]
```

```python
plt.plot(x,y)
plt.plot(x,data1)
plt.plot(x,data2)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,data1, marker="o", ls="--")
plt.plot(x,data2, marker="s", ls="-.")
# 범례(y,data1,data2)
plt.legend(["y","data1","data2"],ncol=3)
plt.show()
```
