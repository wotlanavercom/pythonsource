#### 파이썬에서 예외 처리
    - try ~ except : try(오류 발생 코드) except(예외 발생시 처리 코드)
    - try ~ except 오류명 :
    - try ~ except 오류명 ~ else 오류가 발생하지 않는 경우 처리 코드
    - try ~ except ~ else  ~ finally 오류 여부와 상관없이 무조건 실행 코드

```python
try:
    x = [10,20,30]

    print(x[0])
    print(x[1])
    print(x[3])
except:
    print("에러 발생")
```

```python
names = ["Kim", "Park", "Lee"]

try:
    z = "Let"
    x = names.index(z)
    print("{} Found it!! in name {}".format(z, x+1))
except ValueError:
    print("Not Fount")
```

```python
try:
    number_input = int(input("정수 입력"))
except:
    print("입력값을 확인해 주세요")
else:
    print("원 반지름",number_input)
    print("원 둘레", 2 * 3.14 * number_input)
    print("원 넓이", 3.14 * number_input * number_input)
```

```python
names = ["Kim", "Park", "Lee"]

try:
    z = "Let"
    x = names.index(z)
    print("{} Found it!! in name {}".format(z, x+1))
except Exception:
    print("Not Fount")
```

```python
names = ["Kim", "Park", "Lee"]

try:
    z = "Let"
    x = names.index(z)
    print("{} Found it!! in name {}".format(z, x+1))
except Exception:
    print("Not Fount")
else:
    print("OK!! else!!")
finally:
    print("오류와 상관없이 무조건 실행")
```

```python
try:
    x = [10,20,30]

    print(x[0])
    print(x[1])
    print(x[3])
finally:
    print("무조건 실행")
```

```python
try:
    number_input = int(input("정수 입력"))
except Exception as e:
    print("입력값을 확인해 주세요",e)
else:
    print("원 반지름",number_input)
    print("원 둘레", 2 * 3.14 * number_input)
    print("원 넓이", 3.14 * number_input * number_input)
```

```python
try:
    a = "333"
    if a == "Kim":
        print("OK!!! 허가")
    else:
        raise ValueError
except ValueError:
    print("오류 발생")
except Exception as e:
    print("다른 오류 발생",e)
else:
    print("OK!!")
```
