### [실습2]

```python
df2 = pd.DataFrame(
    {
        "A":["ha","hi","ho","ha","ho"],
        "B" : ["one","two","one","one","two"],
        "Data1" : np.random.randn(5),
        "Data2" : np.random.randn(5),
    }
)
df2
```

```python
# A열 기준으로 그룹
df2.groupby("A")["Data1"].sum()
```

```python
# B열 기준으로 그룹 : Data1,Data2 합계
df2.groupby("B")[["Data1","Data2"]].sum()
```

```python
# sum, mean, std

df2.groupby("B")[["Data1","Data2"]].agg(["sum","mean","std"])
```

```python
# car_seles.xlsx 데이터 프레임 생성

car_df = pd.read_excel("./data/car_sales.xlsx")
car_df
```

```python
# Branch 기준으로 매출액 합계

car_df.groupby(by="Branch")["Ext Price"].sum()
```

```python
# 위의 결과를 데이터프레임으로 생성
branch_df = pd.DataFrame(car_df.groupby(by="Branch")["Ext Price"].sum().rename("매출액"))   # 컬럼이 하나만 있어서 columns안써도 됨
branch_df
```

```python
# branch_df index 해제

car_merge = branch_df.reset_index(inplace=True)
car_merge
```

```python
# merge

car_merge =  pd.merge(car_df,branch_df)
car_merge
```

```python
# 브랜치별 자동차 종류별 점유율?

car_merge["점유율"] = car_merge["Ext Price"] / car_merge["매출액"] * 100
car_merge
```

### transform() : 연산의 결과를 현재 데이터프레임의 인덱스에 맞춰서 돌려줌

```python
car_df["매출액"] = car_df.groupby("Branch")["Ext Price"].transform("sum")
car_df["점유율"] = car_df["Ext Price"] / car_df["매출액"] * 100
```

```python
car_df
```

### [실습4]

```python
df = pd.DataFrame(
    {
        "School": ["Yeonhi","Yeonhi","sungsam","sungsam","sungsam"],
        "Name" : ["Haena","Gisu","Una","Naeun","Ziho"],
        "Math_s" : [92,70,88,92,70]
    }
)
df
```

```python
# school을 기준으로 수학 성적 평균 구하기

df["Avg"] = df.groupby("School")["Math_s"].transform("mean")
```

```python
df
```

```python
# 등급
def grade(num):
    if num > 90:
        grade = "A"
    elif num >80:
        grade = "B"
    elif num > 70:
        grade = "C"
    else:
        grade = "F"
    return grade
```

```python
df["grade"] = df["Math_s"].apply(grade)
```
