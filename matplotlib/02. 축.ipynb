## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title("선그래프", fontdict={"size":20})
```

### 축설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html

#### 1. 이름 설정 
><b>1) xlabel(), ylabel()</b>

```python
plt.plot(x,y)
plt.title("선그래프", fontdict={"size":20})
plt.xlabel("x 축")
plt.ylabel("y 축")
```

><b>2) 이름 및 다양한 설정</b>

matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)

- fontdict : font 설정을 dict 구조로 설정
- labelpad : default 4.0 설정됨(label과 차트 사이의 간격)
- **kwargs(?)
    - color : rgb설정, 색상명 가능
- loc : 축 이름 위치 설정('left', 'center', 'right'),  y 축 (top, bottom, center)

```python
plt.plot(x,y)
plt.title("선그래프", fontdict={"size":20})
plt.xlabel("x 축", color="red", loc="right")
plt.ylabel("y 축", color="#00aa00", loc="top")
```

```python
plt.plot(x,y)
plt.title("chart", fontdict={"size":20})
plt.xlabel("x", labelpad=15, fontdict={"family":"serif","color":"blue","weight":"bold","size":16})
plt.ylabel("y", labelpad=20, fontdict={"family":"serif","color":"deeppink","weight":"bold","size":12})
```

#### 2. 축 눈금 설정 - xticks(), yticks()

```python
# 눈금설정

xticks = [1,2,3]
yticks = [3,6,9,12]

plt.plot(x,y)
plt.xticks(xticks)
plt.yticks(yticks)
plt.show()
```

```python
plt.plot(x,y)
plt.xticks(xticks,fontsize=20)
plt.yticks(yticks,fontsize=14)
plt.show()
```
