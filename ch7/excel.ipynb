```python
from openpyxl import Workbook, load_workbook

wb = Workbook()
ws = wb.active

# 리스트 형태로 데이터 삽입
# 리스트 [] : 아무거나 담을 수 있음

ws.append(["번호", "영어", "수학"])
for i in range(1,11):
    ws.append([i, randint(0,100), randint(0,100)])

# 특정 과목 - 영어 컬럼의 점수 출력
print(ws['B'])

for cell in ws['B']:
    print(cell.value)

for cols in ws['B:C']:
    for cell in cols:
        print(cell.value, end=" ")
    print()

for rows in ws['2:7']:
    for cell in rows:
        print(cell.value, end=" ")
    print()

# iter_rows() : 전체 row 를 가져오는 함수, iter_cols() : 전체 cols 가져오는 함수

wb.save("range.xlsx")
wb.close()
```

```python
# range.xlsx 읽기

wb = load_workbook("range.xlsx")
ws = wb.active

for row in ws.iter_rows(min_row=2):
    if int(row[1].value) > 80:
        print(row[0].value,"번 학생의 영어 점수는 ", row[1].value, "입니다")


# 영어 과목명을 컴퓨터로 변경 후 새로운 엑셀 파일로 저장
for row in ws.iter_rows(max_row=1):
    for cell in row:
        if cell.value == "영어":
            cell.value = "컴퓨터"
wb.save("range_modified.xlsx")
wb.close()
```

```python
# 행, 열 삽입
wb = load_workbook("range.xlsx")
ws = wb.active

# ws.insert_rows(8)

ws.insert_rows(8,5)

ws.insert_cols(2,3)

wb.save("range_insert.xlsx")
wb.close()
```

```python
# 차트 작성

from openpyxl.chart import BarChart,Reference

wb = load_workbook("range.xlsx")
ws = wb.active

# 범위 지정
bar_value = Reference(ws, min_row=1, max_row=11, min_col=2, max_col=3)
# 차트 종류 지정
bar_chart =BarChart()
#차트 데이터 추가 ( titles_from_data=True : 범위에 제목 행이 추가되어 있는 경우)
bar_chart.add_data(bar_value, titles_from_data=True)

ws.add_chart(bar_chart, "E1")

wb.save("range_chart.xlsx")
wb.close()
```

```python
from openpyxl.chart import LineChart,Reference

wb = load_workbook("range.xlsx")
ws = wb.active

# 범위 지정
line_value = Reference(ws, min_row=1, max_row=11, min_col=2, max_col=3)
# 차트 종류 지정
line_chart =LineChart()
#차트 데이터 추가 ( titles_from_data=True : 범위에 제목 행이 추가되어 있는 경우)
line_chart.add_data(bar_value, titles_from_data=True)

line_chart.title = "성적표"
line_chart.style = 20
line_chart.y_axis.title = "점수"
line_chart.x_axis.title = "번호"

ws.add_chart(line_chart, "E1")

wb.save("range_chart.xlsx")
wb.close()
```
