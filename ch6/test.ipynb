#### 파이썬에서 예외 처리
    - try ~ except : try(오류 발생 코드) except(예외 발생시 처리 코드)

```python
try:
    x = [10,20,30]

    print(x[0])
    print(x[1])
    print(x[3])
except:
    print("에러 발생")
```
